```python
import os
from dotenv import load_dotenv

# Load the environment variables from the .env file
load_dotenv()

# Fetch the keys (LangChain will also automatically detect these in the background)
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")

if not GOOGLE_API_KEY or not TAVILY_API_KEY:
    print("⚠️ Warning: One or both API keys are missing. Check your .env file.")
else:
    print("✅ Environment variables loaded successfully!")
```

```python
# LangChain and Agent imports
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.tools.tavily_search import TavilySearchResults

# The new unified agent creator!
from langchain.agents import create_agent
```

```python
from google import genai

client = genai.Client(api_key=GOOGLE_API_KEY)

print("Models available to your key:\n")
for m in client.models.list():
    print(m.name)
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-flash-latest",
    temperature=0
)

print("✅ LLM Initialized")
```

```python
# Setup the Tavily search tool
search_tool = TavilySearchResults(max_results=3)
tools = [search_tool]

print("✅ Tools Ready")
```

```python
# Create the modern Agent
# This completely replaces create_react_agent, AgentExecutor, and hub.pull
agent = create_agent(
    model=llm, 
    tools=tools, 
    system_prompt="You are a helpful, logical assistant. Always search the web for current events or real-time data."
)

print("✅ Agent Created")
```

```python
import time
from google.genai.errors import ClientError

def invoke_with_retry(agent, payload, max_retries=3, base_delay=5):
    for attempt in range(1, max_retries + 1):
        try:
            return agent.invoke(payload)
        except ClientError as e:
            if "RESOURCE_EXHAUSTED" in str(e) and attempt < max_retries:
                wait = base_delay * attempt
                print(f"⚠️ Quota hit, retrying in {wait}s... (attempt {attempt}/{max_retries})")
                time.sleep(wait)
            else:
                raise
```

```python
question = "Who is the current prime minister of Sri Lanka, and what is the current weather in Colombo?"

print(f"🤖 Asking: {question}\n")

try:
    response = invoke_with_retry(agent, {
        "messages": [
            {"role": "user", "content": question}
        ]
    })

    print("\n==================================")
    print("FINAL ANSWER:")
    print("==================================")
    print(response["messages"][-1].content)

except ClientError as e:
    print("❌ Still hitting quota limits after retries.")
    print("→ This means your Google Cloud project's free-tier quota for this model is 0.")
    print("→ Fix: enable billing on the project, or check quotas at")
    print("   https://console.cloud.google.com/iam-admin/quotas")
```
